```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
gender_conv = {"F": 1, "M": 2}
df['sex'] = df['sex'].replace(gender_conv)
addr_conv = {"U": 1, "R": 2}
df['address'] = df['address'].replace(addr_conv)
famsize_conv = {"LE3": 1, "GT3": 2}
df['famsize'] = df['famsize'].replace(famsize_conv)
Pstat_conv = {"T": 1, "A": 2}
df['Pstatus'] = df['Pstatus'].replace(Pstat_conv)
Mjob_conv = {"at_home": 1, "health": 2, "services": 3, "teacher": 4, "other": 5}
df['Mjob'] = df['Mjob'].replace(Mjob_conv)
Fjob_conv = {"at_home": 1, "health": 2, "services": 3, "teacher": 4, "other": 5}
df['Fjob'] = df['Fjob'].replace(Fjob_conv)
rsn_conv = {"course": 1, "home": 2, "reputation": 3, "other": 4}
df['reason'] = df['reason'].replace(rsn_conv)
romRel_conv = {"yes": 1, "no": 2}
df['romantic'] = df['romantic'].replace(romRel_conv)
guard_conv = {"mother": 1, "father": 2, "other": 3}
df['guardian'] = df['guardian'].replace(guard_conv)
schlSup = {"no": 1, "yes": 2}
df['schoolsup'] = df['schoolsup'].replace(schlSup)
famSup = {"no": 1, "yes": 2}
df['famsup'] = df['famsup'].replace(famSup)
paid = {"no": 1, "yes": 2}
df['paid'] = df['paid'].replace(paid)
actv = {"no": 1, "yes": 2}
df['activities'] = df['activities'].replace(actv)
nursry = {"no": 1, "yes": 2}
df['nursery'] = df['nursery'].replace(nursry)
highEd = {"no": 1, "yes": 2}
df['higher'] = df['higher'].replace(highEd)
intrnt = {"no": 1, "yes": 2}
df['internet'] = df['internet'].replace(intrnt)
y = df['G3']
X = df.drop(columns=['school', 'G3'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
regr = LinearRegression()
```

```python
model = regr.fit(X_train, y_train)
```

```python
r_sq = model.score(X_train, y_train)
r_sq
```

```python
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

```python
g1 = input("Enter G1 marks: ")
g2 = input("Enter G2 marks: ")
g1_int = float(g1)
g2_int = float(g2)
```

```python
y_pred = model.predict([[1, 16, 1, 1, 1, 4, 4, 2, 4, 3, 1, 1, 4, 0, 1, 2, 1, 2, 2, 2, 2, 2, 4, 2, 2, 1, 2, 5, 3, g1_int, g2_int]])
```
